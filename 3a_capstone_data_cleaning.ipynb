```python
# Split shoe features into men's and women's.

df['toe_drop_m'] = split_toe_drop[0].map(lambda x: x.replace("Men: ","").replace("mm","") if "Men" in x else 0)
df['toe_drop_w'] = split_toe_drop[1].map(lambda x: x.replace("Women: ","").replace("mm;","") if "Women" in x else 0)

df['heel_height_m'] = split_heel_height[0].map(lambda x: x.replace("Men: ","").replace("mm","") if "Men" in x else 0)
df['heel_height_w'] = split_heel_height[1].map(lambda x: x.replace("Women: ","").replace("mm;","") if "Women" in x else 0)

df['forefoot_height_m'] = split_forefoot_height[0].map(lambda x: x.replace("Men: ","").replace("mm","") if "Men" in x else 0)
df['forefoot_height_w'] = split_forefoot_height[1].map(lambda x: x.replace("Women: ","").replace("mm;","") if "Women" in x else 0)

df['width_m'] = split_width[0].map(lambda x: x.replace("Men: ","") if "Men" in x else 0)
df['width_w'] = split_width[1].map(lambda x: x.replace("Women: ","").replace(";","") if "Women" in x else 0)
```

```python
# Clean up number of star rating percentages.
df["user_5star"] = df["user_5star"].str.replace("%","")
df["user_4star"] = df["user_4star"].str.replace("%","")
df["user_3star"] = df["user_3star"].str.replace("%","")
df["user_2star"] = df["user_2star"].str.replace("%","")
df["user_1star"] = df["user_1star"].str.replace("%","")
```

```python
df
```

```python
# Binarize features.
df['discontinued'] = df['discontinued'].map({'Y': 1, 'N': 0})
df['terrain'] = df['terrain'].map({'Road': 1, 'Trail': 0})
df['use'] = df['use'].map({'Daily running': 1, 'Competition': 0})
df['waterproof'] = df['waterproof'].map({'Waterproof': 1, 'Not waterproof': 0})
df['water_repellent'] = df['water_repellent'].map({'Water repellent': 1, 'Not water repellent': 0})
df['maximalist'] = df['maximalist'].map({'Maximalist': 1, 'Not maximalist': 0})
df['minimalist'] = df['minimalist'].map({'Minimalist': 1, 'Not minimalist': 0})
df['triathlon'] = df['triathlon'].map({'Triathlon': 1, 'Not triathlon': 0})
```

```python
# Change column types.
df = df.astype({
    'core_score': int,
    'user_review_count': int,
    'expert_review_count': int,
    'user_5star': int,
    'user_4star': int,
    'user_3star': int,
    'user_2star': int,
    'user_1star': int,
    'price': int,
    'weight_m': int,
    'weight_w': int,
    'toe_drop_m': int,
    'toe_drop_w': int,
    'heel_height_m': int,
    'heel_height_w': int,
    'forefoot_height_m': int,
    'forefoot_height_w': int
})
```

```python
# Create categorical features.
df['zero_drop_m'] = df['toe_drop_m'].map(lambda x: 1 if x==0 else 0)
df['zero_drop_w'] = df['toe_drop_w'].map(lambda x: 1 if x==0 else 0)
df['low_drop_m'] = df['toe_drop_m'].map(lambda x: 1 if x>=1 and x<=4 else 0)
df['low_drop_w'] = df['toe_drop_w'].map(lambda x: 1 if x>=1 and x<=4 else 0)
```

```python
# Save cleaned data to csv.
df.to_csv('shoe_data_cleaned.csv', index=False)
```
